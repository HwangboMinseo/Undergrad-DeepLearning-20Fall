## **Slicing and dicing**

Slicing and Dicing refers to a way of segmenting, viewing and comprehending data in a database. Large blocks of data is cut into smaller segments and the process is repeated until the correct level of detail is achieved for proper analysis.

```
a = np.arange(10)
print(a)
```

```
##(start:stop:step)
a[2:7:2]
```

```
a[4:]
```

```
a = np.array([[1,2,3],[3,4,5],[4,5,6]]) 
print(a)
```

```
# Returns array of items in the second column 
a[...,1]#모든행의 1열에 있는 것
```

```
# Will slice all items from the second row 
a[1,...]#1행에 있는 모든 것
```

```
# Will slice all items from column 1 onwards 
a[...,1:]#모든 행에서 열1부터 끝까지
```

```
# Slicing using advanced index for column
a[1:3,[1,2]]#행1~2에서 1열,2열
```

# **Broadcast**

넘파이에서 형상이 다른 배열을 계산하기 위해서 지원하는 기능, 쉽고 빠르게 고속으로 계산

```
A = np.array([[36.3, 3, 4.0, 68.0],
             [37.2, 4, 2.0, 77.0],
             [36.8, 2, 0.5, 100.0]])
print(A)
```

```
A.shape
```

```
cal = A.max(axis=0)
print(cal)
```

```
cal.shape
```

```
percentage = 100 * A / cal
print(percentage)
```

##**Random**

np.random.choice

```
np.random.seed(0)
```

```
p = np.array([0.1, 0.15, 0.55, 0.2]) #합치면 1.0 확률과 비슷
```

```
for i in range (10) :
  index = np.random.choice([0,1,2,3], p = p.ravel())
  print( i, " ---- ", index)
```

index = 0 이 뽑힐 확률이 0.1, index =1이 뽑힐 확률이 0.15 ~

$$ P(index = 0) = 0.1, P(index = 1) = 0.15, P(index = 2) = 0.55, P(index = 3) = 0.2.  $$

```
v = [0, 0, 0, 0]
ntest = 1000
for i in range(ntest):
    idx = np.random.choice([0, 1, 2, 3], p = p.ravel())
    v[idx] += 1
```

* (중요) array shuffle하기 (2차원)

학습돌릴때 사진을 셔플돌려서 넣는다. 하지만 사진의 라벨도 사진과 함께 셔플되어야함(사진 라벨을 연결되어 셔플되어야함)

```
x1 = [2, 4, 8, 10, 20] # 2는 0.2와, 4는 0.4와 ..~ 함께 움직여야함
x2 = [0.2, 0.4, -0.8, 1.0, -2.0]
X = np.transpose(np.vstack((x1, x2))) # transpose() -> (2, 5)shape 배열을 (5, 2)로 뒤집음
```

```
print(X)
print(X.shape)
```

```
permutation = np.random.permutation(5)
print(permutation)
```

```
X_shuffle=  X[permutation]
```

```
print(X_shuffle)
```
